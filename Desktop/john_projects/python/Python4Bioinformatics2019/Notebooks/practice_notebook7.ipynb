## How to Get the longest repeat in a sequence
***THe Key is to understand how python reads the strings***

```python
def substring_extra(dna):
    lo,up,sublo,subup = 0,1,0,1
    while up < len(dna):
        if dna[up] != dna [up -1]:
            if subup - sublo < up - lo:
                sublo,subup = lo,up
            lo=up
        up+=1
    if subup - sublo < up - lo:
        sublo,subup = lo,up
    return dna[sublo:subup]
    
    
malaria = "AAAATGCCTTGGCATGACCCCGGATCCCCCCCCCTTC"
print(substring_extra(malaria))
```
